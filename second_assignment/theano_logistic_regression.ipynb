```python
import numpy as np
from theano import *
import theano.tensor as T
rand_gen = np.random

N = 400 # Number of training samples
F = 784 # number of inputs

# generate a dataset:
# D = (input_values, target_class)
D = (rand_gen.randn(N,F), rand_gen.randint(size=N, low=0, high=2) )
training_steps = 10000

# declare theano symbolic variables
# x is input, y is output
x = T.matrix("x")
y = T.vector("y")

# randomly initialize weight vector
# and initialize bias to 0
# (weight and bias vectors are shared
# so that they keep values between
# training iterations)
W = shared(rand_gen.randn(F), name="W")
b = shared(0., name="b")

print "Initial model:"
print "Weights"
print W.get_value()
print "Biases"
print b.get_value()

# construct theano expression graph
p_1 = 1 / (1 + T.exp(-T.dot(x,W) - b)) # probability that target = 1
prediction = p_1 > 0.5 # threshold prediction
loss = -y * T.log(p_1) - (1-y) * T.log(1-p_1) # cross-entropy loss
cost = loss.mean() + 0.01 * (W**2).sum() # cost to minimize
gW, gb = T.grad(cost, [W,b]) #grad wrt weight and bias

# compile
train = function (inputs=[x,y], outputs=[prediction,loss], updates=((W,W-0.1*gW),(b,b -0.1*gb)))
predict = function(inputs=[x], outputs=prediction)
# train
for i in range(training_steps):
    pred, err = train(D[0], D[1])

print "Final model:"
print "Weights"
print W.get_value()
print "Biases"
print b.get_value()
print("target values for D:")
print(D[1])
print("prediction on D:")
print(predict(D[0]))
```
